# Dettecting outliers

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate Q1, Q3, and IQR for each numerical column
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers for each numerical column
outliers = {}
for col in numerical_cols:
    outlier_indices = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])].index
    outliers[col] = outlier_indices

# Display columns with outliers and the number of outliers
print("Columns with potential outliers and the number of outliers:")
for col, indices in outliers.items():
    if len(indices) > 0:
        print(f"- {col}: {len(indices)} outliers")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_with_outliers = ['Duration',  'Tot. Price']

for col in columns_with_outliers:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

# Formatting language column

```python
display(df['Language'].unique())
```

```python
# Format 'Tot. Price' to euros without decimals
df['Tot. Price'] = df['Tot. Price'].apply(lambda x: f"{x:.0f} €")

# Change 'Creation Date' to Spanish date format
df['Creation Date'] = pd.to_datetime(df['Creation Date']).dt.strftime('%d/%m/%Y')

# Replace language codes with country names
language_mapping = {'ES': 'Spain', 'EN': 'UK', 'FR': 'France','DE': 'Germany','NL': 'Netherlands','NO': 'Norway','IT': 'Italy',
                    'DA': 'Denmark','PL': 'Poland','RU': 'Russia','SV': 'Sweden','HR': 'Hungary','FI': 'Finland' }
df['Language'] = df['Language'].replace(language_mapping)

display(df.head())
```

```python
df = df.rename(columns={'Language': 'Country'})
display(df.head())
```

# Average price per day

```python
# Convert 'Tot. Price' back to numeric for calculation if it's not already
if df['Tot. Price'].dtype == 'object':
    df['Tot. Price_numeric'] = df['Tot. Price'].str.replace(' €', '').astype(float)

# Calculate price per day
df['Price per Day'] = df['Tot. Price_numeric'] / df['Duration']

# Group by 'Country' and calculate the average price per day
avg_price_per_day_per_country = df.groupby('Country')['Price per Day'].mean().reset_index()

# Rename the columns for clarity and round the average price
avg_price_per_day_per_country.columns = ['Country', 'Average Price per Day']
avg_price_per_day_per_country['Average Price per Day Rounded'] = avg_price_per_day_per_country['Average Price per Day'].round().astype(int)

# Sort by average price per day in descending order
avg_price_per_day_per_country = avg_price_per_day_per_country.sort_values(by='Average Price per Day', ascending=False)

# Drop the original unrounded column
avg_price_per_day_per_country = avg_price_per_day_per_country.drop(columns=['Average Price per Day'])

# Display the result
display(avg_price_per_day_per_country)
```

We can see a difference between the different segments created by the preferences of each country. This gives us insights to segment based on nationality, as we have an idea of ​​their consumer trends, for example:

*Norway is a premium segment that pays for expensive accommodations, so we can
implement marketing strategies focused on this segment, such as conducting market research on what Norwegians look for in the accommodations they rent or their booking method preferences.

*UK prefer affordable accommodations. We can also use these insights to implement marketing strategies such as discounts and promotions for these segments.

## Average stay duration per country

# Number of reservations per country

```python
# Count the number of reservations per country
reservations_per_country = df['Country'].value_counts().reset_index()

# Rename the columns for clarity
reservations_per_country.columns = ['Country', 'Number of Reservations']

# Sort by the number of reservations in descending order
reservations_per_country = reservations_per_country.sort_values(by='Number of Reservations', ascending=False)

# Display the result
display(reservations_per_country)
```

We can see the top three countries with the most bookings. We can intuit that this is stable demand, and we can adjust prices by raising them for the types of properties these clients most often rent. We can focus our marketing efforts on attracting more clients from Norway, Poland, or the Netherlands, who pay more for their stays.

# Average duration per country

```python
# Group by 'Country' and calculate the total duration
total_duration_per_country = df.groupby('Country')['Duration'].sum().reset_index()

# Rename the duration column to 'Total Duration'
total_duration_per_country.rename(columns={'Duration': 'Total Duration'}, inplace=True)

# Merge with reservations per country data to get the number of reservations
total_duration_per_country = pd.merge(total_duration_per_country, reservations_per_country[['Country', 'Number of Reservations']], on='Country', how='left')

# Calculate the average duration by dividing total duration by number of reservations
total_duration_per_country['Average Duration'] = total_duration_per_country['Total Duration'] / total_duration_per_country['Number of Reservations']

# Round the 'Average Duration' column
total_duration_per_country['Average Duration'] = total_duration_per_country['Average Duration'].round().astype(int)

# Sort by average duration in descending order
total_duration_per_country = total_duration_per_country.sort_values(by='Average Duration', ascending=False)

# Drop the 'Total Duration' and 'Number of Reservations' columns
total_duration_per_country = total_duration_per_country.drop(columns=['Total Duration', 'Number of Reservations'])

# Display the result
display(total_duration_per_country)
```

We can see the countries with the longest stays. As vacation rentals, we're interested in longer stays. We can develop strategies to find more long-stay guests and retain existing guests with promotional offers and adapting the properties these guests typically rent to ensure they're equipped with the necessary features for their needs, such as fast Wi-Fi (since many of these guests work remotely) and monthly payments (since these reservations are large).
We've seen that the Netherlands appears among the highest-paying countries, so we can implement customer service in the Netherlands with the help of customized chatboxes and advertising campaigns in this language.

# Average booking lead time

```python
# Convert date columns to datetime objects
df['Creation Date'] = pd.to_datetime(df['Creation Date'], format='%d/%m/%Y')
df['Reservation Start'] = pd.to_datetime(df['Reservation Start'])

# Calculate booking lead time
df['Booking Lead Time'] = (df['Reservation Start'] - df['Creation Date']).dt.days

# Group by 'Country' and calculate the average booking lead time
avg_lead_time_per_country = df.groupby('Country')['Booking Lead Time'].mean().reset_index()

# Rename the columns for clarity
avg_lead_time_per_country.columns = ['Country', 'Average Booking Lead Time (days)']

# Round the 'Average Booking Lead Time (days)' column
avg_lead_time_per_country['Average Booking Lead Time (days)'] = avg_lead_time_per_country['Average Booking Lead Time (days)'].round().astype(int)

# Sort by average lead time in descending order
avg_lead_time_per_country = avg_lead_time_per_country.sort_values(by='Average Booking Lead Time (days)', ascending=False)

# Display the result
display(avg_lead_time_per_country)
```

We can see that some nationalities prefer to rent in advance, so we can start our campaigns in relation to high season, well before it begins. For example, we can run campaigns focused on the Dutch market, which begins in February, with promotions and marketing focused on their language and preferences.

# Top country count per type of accommodation

```python
# Group by 'Type' and 'Country' and count the occurrences
type_country_counts = df.groupby(['Type', 'Country']).size().reset_index(name='Reservation Count')

# Calculate the total reservation count for each 'Type'
total_reservations_per_type = type_country_counts.groupby('Type')['Reservation Count'].transform('sum')

# Calculate the percentage of reservation count within each 'Type'
type_country_counts['Percentage within Type (%)'] = (type_country_counts['Reservation Count'] / total_reservations_per_type) * 100

# Sort within each 'Type' group by 'Reservation Count' in descending order
type_country_counts_sorted = type_country_counts.sort_values(by=['Type', 'Reservation Count'], ascending=[True, False])

# Get the top 1 country for each 'Type'
top_countries_per_type = type_country_counts_sorted.groupby('Type').head(1).reset_index(drop=True)

# Display the result
display(top_countries_per_type)
```

We can see that small apartments tend to be rented mostly by Spanish residents, so we can name these apartments with catchy Spanish names and encourage these types of guests to leave positive reviews. On the other hand, medium-sized apartments are more frequented by people from the UK, so we can implement English names and adapt these accommodations to their needs, such as having a kettle or details that this segment appreciates, so they leave positive reviews. Finally, large homes are frequented by more than 50% of French residents, so we can use French names for these homes and study their spending habits and preferences to improve their experience and encourage repeat visitors.

# Number of reservations per creation month

```python
# Ensure 'Creation Date' is in datetime format
df['Creation Date'] = pd.to_datetime(df['Creation Date'])

# Extract the month and year from 'Creation Date'
df['Creation Month'] = df['Creation Date'].dt.month
df['Creation Year'] = df['Creation Date'].dt.year

# Group by year and month and count the number of reservations
reservations_per_month_year = df.groupby(['Creation Year', 'Creation Month']).size().reset_index(name='Number of Reservations')

# Calculate the total number of reservations
total_reservations = reservations_per_month_year['Number of Reservations'].sum()

# Calculate the percentage of total reservations
reservations_per_month_year['Percentage of Total (%)'] = (reservations_per_month_year['Number of Reservations'] / total_reservations) * 100

# Round the percentage column
reservations_per_month_year['Percentage of Total (%)'] = reservations_per_month_year['Percentage of Total (%)'].round(2)


# Sort by the number of reservations in descending order
reservations_per_month_year = reservations_per_month_year.sort_values(by='Number of Reservations', ascending=False)

# Display the result
display(reservations_per_month_year)
```

We can see that from March to May more than 50% of the reservations are created compared to the high season, so we can start our advertising campaigns during this period, focusing our efforts and attention on the entry of reservations.

# Top countries per creation month

```python
# Ensure 'Creation Date' is in datetime format
df['Creation Date'] = pd.to_datetime(df['Creation Date'])

# Extract the month and year from 'Creation Date'
df['Creation Month'] = df['Creation Date'].dt.month
df['Creation Year'] = df['Creation Date'].dt.year

# Group by year and month and count the number of reservations
reservations_per_month_year = df.groupby(['Creation Year', 'Creation Month']).size().reset_index(name='Number of Reservations')

# Calculate the total number of reservations
total_reservations = reservations_per_month_year['Number of Reservations'].sum()

# Calculate the percentage of total reservations
reservations_per_month_year['Percentage of Total (%)'] = (reservations_per_month_year['Number of Reservations'] / total_reservations) * 100

# Round the percentage column
reservations_per_month_year['Percentage of Total (%)'] = reservations_per_month_year['Percentage of Total (%)'].round(2)


# Group by year, month, and country and count the number of reservations
reservations_by_month_year_country = df.groupby(['Creation Year', 'Creation Month', 'Country']).size().reset_index(name='Reservation Count')

# Find the country with the maximum reservation count for each month and year
top_country_per_month_year = reservations_by_month_year_country.loc[reservations_by_month_year_country.groupby(['Creation Year', 'Creation Month'])['Reservation Count'].idxmax()]

# Select only the relevant columns for merging
top_country_per_month_year = top_country_per_month_year[['Creation Year', 'Creation Month', 'Country']]

# Merge this information back with the reservations per month year table
reservations_per_month_year = pd.merge(reservations_per_month_year, top_country_per_month_year, on=['Creation Year', 'Creation Month'], how='left')

# Rename the 'Country' column to 'Top Country' for clarity in the merged table
reservations_per_month_year.rename(columns={'Country': 'Top Country'}, inplace=True)

# Sort by the number of reservations in descending order
reservations_per_month_year = reservations_per_month_year.sort_values(by='Number of Reservations', ascending=False)


# Display the result
display(reservations_per_month_year)
```

We can see that more than 50% of reservations are created from March to May compared to the peak season, so we can begin our advertising campaigns during this period, focusing our efforts and attention on incoming reservations.
The majority of these clients are Spanish, so we can adapt our campaigns to their preferences, taking into account the aforementioned fact that they prefer smaller accommodations and shorter stays.
